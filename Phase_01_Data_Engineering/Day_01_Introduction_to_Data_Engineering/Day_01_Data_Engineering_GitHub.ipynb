

> 5. Filtering Rows



```python
cs_students = df[df["department"]=="Computer Science"]   # select students from the "Computer Science" department
high_attendance = df[df["attendance_percentage"] > 90]   # filters where attendance is above 90
```

 ****

> 6. Creating a new column



```python
df['total_score']=df['math_score']+df['english_score']+df['programming_score']+df['science_score']   # adding all score to get total score
```

```python
# Finds the student with the highest total score and stores that row in top_student.

top_student = df.loc[df['total_score'].idxmax()]     # df.loc[]  ->  Used to select a row using index label.
print(top_student)                                   # .idxmax() ->  Finds the index (row number) of the highest value.

```



> 7. Grouping and Aggregation.



```python
dept_avg = df.groupby('department')['total_score'].mean().round(2)   # Average of total_score for each department
print(dept_avg)
```



> 8. Sort Data


```python
df_sorted = df.sort_values(by='total_score',ascending=False)
print("="*50)
print("        Top 10 Students by total score:")
print("="*50)
print(df_sorted[['name','department','total_score']].head(10))
```

# Practice Questions

```python
""" Question 1:
How many students are from the Electronics department?"""

students_in_ec = df[df["department"]=="Electronics"]
print("Total Number of Students in Electronics:")
print(len(students_in_ec))
```

```python
""" Question 2:
What is the average attendance of female students?"""

female_students = df[df['gender']=="Female"]
print("Average attendance of female students:")
avg_attendance = female_students["attendance_percentage"].mean()
print(avg_attendance)
```

```python
""" Question 3:
Who is the student with the lowest programming score?"""

df_sort_prog_score = df.sort_values(by="programming_score",ascending= True)
print("The student with the lowest programming score")
print(df_sort_prog_score['name'].head(1))
```

```python
"""Question 4:
How many students scored above 80 in math?"""

df_math_score = df[df["math_score"]>80]
print("Students scored above 80 in math")
print(len(df_math_score))
```

```python
""" Question 5:
What is the average total_score for students with attendance above 90%?"""

attendance_above_90 = df[df["attendance_percentage"]>90]
print("The average total_score for students with attendance above 90%")
print(attendance_above_90["total_score"].mean())
```

# Student Data Explorer -MINI PROJECT 1

## Objective
Analyze student performance data using Pandas and generate key insights.

## Dataset Columns
- name
- gender
- department
- math_score
- programming_score
- attendance_percentage
- total_score

## Tasks

### 1. Dataset Overview
- Total students
- Total columns
- Unique departments

### 2. Department Analysis
- Student count by department

### 3. Score Analysis
- Highest, lowest, and average math score
- Average programming score
- Average attendance

### 4. Gender Analysis
- Average total score by gender

### 5. Top Performers
- Top 5 students by total score

### 6. Attendance Analysis
- Students with attendance below 75%


Coding:

```python

print("="*60)
print("                  Student Data Explorer                                           ")
print("="*60)
print("[1] Dataset Objective:\n---------------------------------------------")
print("Total Number of students: ",df.shape[0])
print("Total Number of columns: ",df.shape[1])
print("Total Number of unique departments: ",len(df["department"].unique()))
print(df['department'].unique(),"\n")
print("[2] Department-wise Analysis:\n---------------------------------------------")
print("The count of students in each department:\n",df['department'].value_counts(),"\n")

print("[3] Score Analysis:\n---------------------------------------------")
 # Highest math score
print("The Highest math score:",df["math_score"].max())

# Lowest math score
print("The Lowest math score:",df["math_score"].min())

# Average math score
print("The Average math score:",df["math_score"].mean().round(2))

# Average Programming score
print("The Average Programming score:",df["programming_score"].mean().round(2))

# Average attendance percentage
print("The Average attendance percentage:",df["attendance_percentage"].mean().round(2),"\n")

print("[4] Gender wise Analysis:\n---------------------------------------------")
# Average total score grouped by gender
total_score_by_gender = df.groupby('gender')['total_score'].mean().round(2)
print("The Average total score grouped by gender:\n",total_score_by_gender,"\n")
print("[5] Top performers:\n---------------------------------------------")
# The top 5 students based on total_Score.
top_5_students = df.sort_values(by='total_score',ascending=False).head(5)
print("The Top 5 students based on total_Score:\n",top_5_students[['name','department','total_score']],"\n")

print("[6] Attendance Analysis:\n---------------------------------------------")
# Students whose attendance is below 75%
stud_below_75 = df[df['attendance_percentage']<75]
if len(stud_below_75)>0:
  print("The Students whose attendance is below 75%:\n",stud_below_75[['name','department','total_score','attendance_percentage']])
else:
  print("All students have attendance above 75%")

```
